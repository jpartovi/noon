```python
import sys
sys.path.insert(0, '/Users/anika/noon/noon-agent')

from noon_agent.helpers import inline_schema_refs
from noon_agent.schemas import ParsedIntent
import json

# Original schema (will have $refs)
original = ParsedIntent.model_json_schema()
print("ORIGINAL (with $ref):")
print(json.dumps(original, indent=2))

print("\n" + "="*80 + "\n")

# Fixed schema (should NOT have $refs)
fixed = inline_schema_refs(original)
print("FIXED (no $ref):")
print(json.dumps(fixed, indent=2))

# Check if $ref exists
if "$ref" in json.dumps(fixed):
  print("\n❌ ERROR: Fixed schema still contains $ref!")
else:
  print("\n✅ SUCCESS: Fixed schema has no $ref!")

if "$defs" in fixed:
  print("❌ ERROR: Fixed schema still contains $defs!")
else:
  print("✅ SUCCESS: Fixed schema has no $defs!")
```
